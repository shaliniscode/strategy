```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from datetime import date, timedelta
```

```python
start_date = date(2025, 1 ,13)
end_date = date(2025, 1, 22)
```

```python
num_days = (end_date - start_date).days
date_list = [start_date + timedelta(days= i) for i in range(num_days + 1)]
```

```python
for single_date in date_list:
    print(single_date)
```

```python
temp = [36.31, 36.37, 36.38, 36.38, 36.29, 36.28, 36.12, 36.38, 36.37,36.39,36.38 ]
```

```python
from decimal import Decimal
```

```python
body_temp =[]
start = Decimal('36.20')
end = Decimal('36.50')
step = Decimal('0.01')
while start < end:
    body_temp.append(start)
    start += step
```
